```python
# Tiny numeric demo
import numpy as np
X = np.linspace(0,1,10)
y = X*2.0 + 0.1
def predict_linear(x, slope=2.0, intercept=0.1):
    return x*slope + intercept
pred = predict_linear(X)
print('MSE:', float(((pred-y)**2).mean()))
```
